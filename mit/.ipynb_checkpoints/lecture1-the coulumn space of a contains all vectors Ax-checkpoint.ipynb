```python
%matplotlib widget
from matplotlib import pyplot as plt
import numpy as np


fig = plt.figure()

# add axes
ax = fig.add_subplot(111,projection='3d')

xx, yy = np.meshgrid(range(-10, 10), range(-10, 10))
x = 2*xx + 3*yy
y = 2*xx + 4*yy
z = 3*xx + 7*yy

# plot the plane
ax.plot_surface(x, y, z, alpha=0.5)
ax.quiver(0, 0, 0, 50, 50, 50)
ax.scatter(0, 0, 0)

# plt.show()
```
